```python
# Configuración inicial
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
import os
import time

# Configurar el modelo
llm = ChatOpenAI(
    base_url=os.getenv("OPENAI_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="gpt-4.1",
    temperature=0.1  # Baja temperatura para razonamiento más consistente
)

print("✓ Modelo configurado para Chain-of-Thought")
print("✓ Temperature baja para razonamiento consistente")
```

```python
# Comparación directa: razonamiento directo vs CoT
# Proyecto: Sistema de Servicios de Vulcanización

def comparar_sin_vs_con_cot():
    print("=== COMPARACIÓN: SIN CoT vs CON CoT ===")
    print("=== Contexto: Taller de Vulcanización ===")
    
    # Problema complejo del contexto de vulcanización
    problema = """Un cliente llega al taller de vulcanización con su auto.
    Necesita los siguientes servicios:
    - Cambio de 2 neumáticos delanteros: $45.000 c/u
    - Reparación de pinchazo en neumático trasero: $8.000
    - Alineación y balanceo completo: $25.000
    
    El taller tiene una promoción vigente: 'Si el total supera $100.000, 
    se aplica un 15% de descuento en el servicio más caro'.
    
    ¿Cuánto paga el cliente en total?"""
    
    # Prompt sin CoT
    prompt_sin_cot = f"""Eres un asistente de un taller de vulcanización. Resuelve este caso:
    
{problema}
    
Respuesta:"""
    
    # Prompt con CoT
    prompt_con_cot = f"""Eres un asistente experto de un taller de vulcanización. 
Resuelve este caso paso a paso:
    
{problema}
    
Piensa paso a paso:
1. Lista todos los servicios solicitados y sus precios
2. Calcula el subtotal sin descuento
3. Verifica si aplica la promoción (¿supera $100.000?)
4. Identifica el servicio más caro para aplicar el descuento
5. Calcula el descuento y el total final
    
Razonamiento:"""
    
    # Probar sin CoT
    print("\n1. SIN CHAIN-OF-THOUGHT:")
    print("-" * 40)
    try:
        response_sin = llm.invoke([HumanMessage(content=prompt_sin_cot)])
        print(response_sin.content)
    except Exception as e:
        print(f"Error: {e}")
    
    print("\n" + "="*60)
    
    # Probar con CoT
    print("\n2. CON CHAIN-OF-THOUGHT:")
    print("-" * 40)
    try:
        response_con = llm.invoke([HumanMessage(content=prompt_con_cot)])
        print(response_con.content)
    except Exception as e:
        print(f"Error: {e}")
    
    print("\n=== ANÁLISIS - Aplicado a Vulcanización ===")
    print("• Sin CoT: Puede calcular mal descuentos o saltarse la promoción")
    print("• Con CoT: Verifica cada servicio y aplica promociones correctamente")
    print("• CoT es clave para: cotizaciones, descuentos y diagnósticos multi-servicio")

# Ejecutar comparación
comparar_sin_vs_con_cot()
```
